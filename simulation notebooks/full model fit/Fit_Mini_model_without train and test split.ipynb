```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
```

```python
tripdata = pd.read_csv("../../dataset/final_bike_sharing.csv")
```

```python
tripdata.head(3)
```

```python
tripdata.columns
```

```python
cols = ['tripduration','dist','hour','birthyear','temp','snowdepth']
```

```python
df = tripdata[cols]
```

```python
data = df.head(200)
```

```python
data
```

```python
X = data.drop('tripduration', axis=1)
y = data['tripduration']
model_lr = LinearRegression(fit_intercept=True).fit(X, y)
```

```python
print("Intercept is", model_lr.intercept_)
print("coeffients are ", model_lr.coef_)
```

```python
y_pred = model_lr.predict(X)
rmse = mean_squared_error(y, y_pred, squared=False)
print("RMSE is of linear regression is = ", rmse)
print("R2 is of linear regression is = ", r2_score(y, y_pred))
```
