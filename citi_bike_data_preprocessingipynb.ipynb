## CitiBike data is collected for 1 year 2018
https://ride.citibikenyc.com/system-data

Data Schema:
Trip Duration (seconds)

*   Start Time and Date
*   Stop Time and Date
*   Start Station Name
*   End Station Name
*   Station ID
*   Station Lat/Long
*   Bike ID
*   User Type (Customer = 24-hour pass or 3-day pass user; Subscriber = Annual    Member)
*   Gender (Zero=unknown; 1=male; 2=female)
*   Year of Birth
*   name
*   Short name
*   

```python
import pandas as pd
```

```python
tripdata = pd.read_csv("/content/JC-201801-citibike-tripdata.csv")
tripdata.head(10)
```

```python
tripdata.count()
```

```python
tripdata.drop_duplicates()
```

```python
tripdata = tripdata.rename(columns={"start station id": "station_id"})
tripdata.head(2)
```

```python
stations_data = pd.read_csv("/content/citibike-stations.csv")
stations_data.head(2)
```

```python
stations_data.isnull().sum()
```

```python
df = pd.merge(tripdata,stations_data, on="station_id", how="left")
```

```python
df.head(2)
```

```python
df.count()
```

```python
df.isnull().sum()
```

```python
na = df[df['name'].isna()]
```

```python
na.groupby('station_id').count()
```
